```python

import tensorflow as tf
# import random
import math
from tensorflow import keras
from Util import readBase,trata_tf_palavra,trata_tf_tf_idf,trata_tf_3
```

```python
(train_data, train_labels), (test_data, test_labels) = trata_tf_tf_idf(readBase('colecao_dourada_2_class_unbalanced.csv'), 0.5)
```

```python
vocab_size = 2267

train_data = keras.preprocessing.sequence.pad_sequences(train_data,
                                                        value=0,
                                                        padding='post',
                                                        maxlen=vocab_size)

test_data = keras.preprocessing.sequence.pad_sequences(test_data,
                                                       value=0,
                                                       padding='post',
                                                       maxlen=vocab_size)
```

```python
model = keras.Sequential()
model.add(keras.layers.Embedding(vocab_size, 64))
model.add(keras.layers.GlobalAveragePooling1D())

#model.add(keras.layers.LSTM(32,recurrent_activation='sigmoid'))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(32, activation=tf.nn.relu6))
model.add(keras.layers.Dense(16, activation=tf.nn.relu))
model.add(keras.layers.Dense(1, activation=tf.nn.sigmoid))

model.summary()

model.compile(optimizer=tf.train.AdamOptimizer(),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
limit = 45
x_val = train_data[:limit]
partial_x_train = train_data[limit:]


y_val = train_labels[:limit]
partial_y_train = train_labels[limit:]


history = model.fit(partial_x_train,
                    partial_y_train,
                    epochs=4,
                    batch_size=1,
                    validation_data=(x_val, y_val),
                    verbose=1)
```

```python
results = model.evaluate(test_data, test_labels)
```
